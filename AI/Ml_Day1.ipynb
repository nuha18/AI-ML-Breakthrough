# **AI/ML Day 1 TASK**

```python
import random

class Pet:
  def __init__(self, name, age, species):
    self.name = name
    self.age = age
    self.species = species
  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Species: {self.species}")

class Dog(Pet):
  def __init__(self, name, age, breed, color):
    super().__init__(name, age, "Dog")
    self.breed = breed
    self.color = color
  def display_info(self):
    super().display_info()
    print(f"Breed: {self.breed}")
    print(f"Color: {self.color}\n")

class Cat(Pet):
  def __init__(self, name, age, breed, color):
    super().__init__(name, age, "Cat")
    self.breed = breed
    self.color = color
  def display_info(self):
    super().display_info()
    print(f"Breed: {self.breed}")
    print(f"Color: {self.color}\n")

class Rabbit(Pet):
  def __init__(self, name, age, size, breed, color):
    super().__init__(name, age, "Rabbit")
    self.size = size
    self.breed = breed
    self.color = color
  def display_info(self):
    super().display_info()
    print(f"Size: {self.size}")
    print(f"Breed: {self.breed}")
    print(f"Color: {self.color}\n")

# Dictionaries for storing pets
pets = {}
# Function for adding pets with random id
def add_pet(pet):
  pet_id = random.randint(1000, 9999)
  pets[pet_id] = pet
  print(f"Pet added with ID: {pet_id}")

# Function for adopting pets
def adopt_pet(pet_id):
  if pet_id in pets:
    adopted_pet = pets.pop(pet_id)
    print(f"Congratulations you have adopted a {adopted_pet.species} named {adopted_pet.name}")
  else:
    print("Pet not found: Invalid ID")

# Tuple to store pet preferences
pet_preferences = {
    "Dog": ("Bones", "playing with a ball"),
    "Cat": ("Fish", "playing with laser pointer"),
    "Rabbit": ("Carrots", "harness walking")
}
def show_pet_preferences(species):
  if species in pet_preferences:
    print(f"Pet Preferences for {species}:")
    print(f"Favorite Food: {pet_preferences[species][0]}")
    print(f"Favorite Activity: {pet_preferences[species][1]}")
  else:
    print("No pet preferences found for this species.")

# Function to view the pets available
def view_pets():
  if not pets:
    print("No pets available for adoption.")
    return
  for pet_id, pet in pets.items():
    print(f"Pet ID: {pet_id}")
    pet.display_info()
    if pet.species in pet_preferences:
      show_pet_preferences(pet.species)
      print("\n")

# Menu for user to view, add and adopt pets
def show_menu():
  while True:
    print("\nPet Adoption System Menu:")
    print("1. View Pets")
    print("2. Add Pet")
    print("3. Adopt Pet")
    print("4. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
      view_pets()
    elif choice == "2":
      species = input("Enter pet species: ")
      name = input("Enter pet name: ")
      age = int(input("Enter pet age: "))
      if species.lower() == "dog":
        breed = input("Enter dog breed: ")
        color = input("Enter dog color: ")
        pet = Dog(name, age, breed, color)
      elif species.lower() == "cat":
        breed = input("Enter cat breed: ")
        color = input("Enter cat color: ")
        pet = Cat(name, age, breed, color)
      elif species.lower() == "rabbit":
        size = input("Enter rabbit size: ")
        breed = input("Enter rabbit breed: ")
        color = input("Enter rabbit color: ")
        pet = Rabbit(name, age, size, breed, color)
      else:
        print("Invalid species. Please choose from Dog, Cat, or Rabbit.")
        continue
      add_pet(pet)
    elif choice == "3":
      pet_id = int(input("Enter the pet ID to adopt: "))
      adopt_pet(pet_id)
    elif choice == "4":
      print("Thank you for using the Pet Adoption System. Exiting! Goodbye!")
      break
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
    show_menu()
```
